```python
from utils.data import load_data
from utils.visualization import visualize_explanation
from torch_geometric.utils import to_dense_adj
```

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
dataset=load_data(with_gt=False)
```

```python
observation_dict=[]
for i in range(len(dataset)):
    data = dataset[i]
    adj = to_dense_adj(data.edge_index)[0]
    
    motif_adj = adj[-5:, -5:]
    
    num_internal_edges = int(motif_adj.sum().item() / 2)
    observation_dict.append({data.y.item(),num_internal_edges})
    if data.y.item()==0 and num_internal_edges!=5:
        print("an error")
    elif data.y.item()==1 and num_internal_edges!=6:
        print("an error")

```
