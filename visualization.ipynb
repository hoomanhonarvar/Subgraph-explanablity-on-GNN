```python
from torch_geometric.datasets import BA2MotifDataset
from torch_geometric.utils import to_networkx
import networkx as ntx
import matplotlib.pyplot as plt
import pickle
from torch.utils.data import Subset
```

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
def visualize_graph(data,motif_nodes=None,title="BA2Motif",save_path=None):
    G=to_networkx(data,to_undirected=True)
    pos=ntx.spring_layout(G,seed=404131029)

    node_colors=[]
    for node in G.nodes():
        if motif_nodes is not None and node in motif_nodes:
            node_colors.append('red')
        else:
            node_colors.append('blue')

    ntx.draw(G,pos,node_color=node_colors,with_labels=True)
    plt.title(title)
    if save_path is not None:
        plt.savefig(save_path)
    plt.show()
    plt.close()
    return G
```

```python
dataset=BA2MotifDataset(root='./data/BA2Motif')
data=dataset[20]
print(data.keys)
visualize_graph(data)
```

```python
data=dataset[500]
print(data.keys)
visualize_graph(data)
```
